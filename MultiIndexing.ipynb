```python
array = objects.values.T.tolist()
```

```python
array
```

```python
array_index = array[1:3]
```

```python
array_index
```

```python
index = pd.MultiIndex.from_arrays(array_index, names = ['Shape','color'])
```

```python
MultiIndexDataframe = pd.DataFrame({'Object':array[0], 'Quantity':array[3]}, index = index)
```

```python
MultiIndexDataframe

```

```python
MultiIndexDataframe.groupby(level = ['color']).sum()
```

```python
#how to call a sub division from a multi index DF
MultiIndexDataframe.groupby(by = ['Shape','color']).get_group(('rectangle','white'))
```

```python
# how to groupby with a column which is not in multi indexing

MultiIndexDataframe.groupby([pd.Grouper(level = 'Shape'),'Object']).sum()
```
